```python
import torch
import torch.nn as nn
import torch.optim as optim

device = "cuda" if torch.cuda.is_available() else "cpu"

# 1. RE-INITIALIZE OUR MODEL ARCHITECTURE
class SimpleMLP(nn.Module):
    def __init__(self):
        super().__init__()
        self.hidden_layer = nn.Linear(in_features=4, out_features=8)
        self.output_layer = nn.Linear(in_features=8, out_features=1)
        self.relu = nn.ReLU()

    def forward(self, x):
        x = self.hidden_layer(x)
        x = self.relu(x)
        x = self.output_layer(x)
        return x

model = SimpleMLP().to(device)

# 2. DEFINE THE TEACHER (LOSS) AND THE MECHANIC (OPTIMIZER)
# Mean Squared Error tells us: How far away is our guess from the target?
loss_function = nn.MSELoss()

# Adam will tweak our model's parameters. lr=0.1 is the step size (learning rate).
optimizer = optim.Adam(model.parameters(), lr=0.1)

# 3. CREATE TOY DATA AND A TARGET ANSWER
toy_input = torch.tensor([[0.5259, 2.0778, 0.1717, 0.8955]]).to(device)
true_target = torch.tensor([[350.0]]).to(device) # The real answer we want to hit!

print("--- TRAINING PROCESS ---")

# Let's run a loop to train the network over 50 steps (epochs)
for epoch in range(1, 51):
    # Step A: Clear out past memory adjustments
    optimizer.zero_grad()

    # Step B: Make a guess (Forward Pass)
    prediction = model(toy_input)

    # Step C: Calculate how wrong the guess was
    loss = loss_function(prediction, true_target)

    # Step D: Backpropagation (Calculate exactly how to turn the knobs)
    loss.backward()

    # Step E: Tweak the weights based on the calculations
    optimizer.step()

    # Print progress every 10 steps
    if epoch % 10 == 0 or epoch == 1:
        print(f"Step {epoch:02d} | Current Guess: {prediction.item():.4f} | Loss (Error Score): {loss.item():.4f}")

print("\n--- FINAL TRAINING RESULT ---")
print(f"Target Answer: {true_target.item()}")
print(f"Model Final Guess: {prediction.item():.4f}")
```
