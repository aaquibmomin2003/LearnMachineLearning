```python
import torch
import torch.nn as nn

device = "cuda" if torch.cuda.is_available() else "cpu"
print(f"Using Device Acceleration: {device.upper()}")

# 1. DEFINE THE BRAIN STRUCTURE (Watch the indentations here!)
class SimpleMLP(nn.Module):
    def __init__(self):
        super().__init__()
        # These belong to __init__
        self.hidden_layer = nn.Linear(in_features=4, out_features=8)
        self.output_layer = nn.Linear(in_features=8, out_features=1)
        self.relu = nn.ReLU()

    def forward(self, x):
        # These belong to forward, aligned perfectly under __init__
        x = self.hidden_layer(x)
        x = self.relu(x)
        x = self.output_layer(x)
        return x

# 2. INITIALIZE AND RUN THE MODEL
model = SimpleMLP().to(device)
print("\n--- MODEL STRUCTURE ARCHITECTURE ---")
print(model)

# Create a random 4-number input tensor and move it to the GPU
toy_input = torch.randn(1, 4).to(device)

# Pass the data forward through the model
prediction = model(toy_input)

print("\n--- INPUT DATA ---")
print(toy_input.cpu().detach().numpy())

print("\n--- FINAL MODEL PREDICTION ---")
print(prediction.cpu().detach().numpy())
```
